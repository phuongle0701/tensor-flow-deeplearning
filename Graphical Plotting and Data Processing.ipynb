##  Graphics in Python:

### Package `Matplotlib`: 

* `Matplotlib`is a functional interface similar to Matlab via the `pyplot`module for simple interactive use, as well as an object-oriented interface that is useful for more complex graphic creations

```python
import warnings
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
import numpy as np
```

## Customize the plots: 

```python
fig = plt.figure(figsize = (10,4))
ax = plt.subplot(1,1,1)

plt.plot(x,np.cos(x), color = 'red', linewidth = 2, 
        linestyle = 'dashed', label = 'cosine curve')

plt.plot(x, np.sin(x), color = 'blue', linewidth = 2, 
        linestyle = '-', 
        label = 'sine curve')
### Styling the plot: 
plt.legend(loc = 'best', fontsize = 14)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Sine and Cosine Plots')
pass
```

### Plots layouts:

```python
fig, axes = plt.subplots(2,2, figsize = (8,8))

axes[0,0].plot(x,y, 'r')
axes[0,1].plot(x,y,'g')
axes[1,0].plot(x,y, 'b')
axes[1,1].plot(x,y,'k')

for ax in axes.ravel(): 
    ax.margins(0.05)
pass
```

## Functions to draw linear regression models: 

* 2 main faunctions in seaborn to visualize the linear relationship. These functions are `regplot()`and `lmplot(). 

* In the simplest invocation, both functions draw a scatteprlot of x and y and then fit the regression model `y ~ x`and plot the resulting regression line, and a 95% confidence interval for that regression. 

```python
sns.regplot(x = 'total_bill', y = 'tip',
           data = tips)
pass
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
pass
```

## Jittering the points:

```python
sns.lmplot(x = 'size', y ='tip', data = tips)
pass
```

```python
## Now we add the jitter argument: 

sns.lmplot(x = 'size', y = 'tip', 
           data = tips, x_jitter = 0.1)
pass
```

* The second option is to collapse over the observations in each discrete bin to plot the estimate of central tendency along with a confidence interval

```python
sns.lmplot(x = 'size', y = 'tip', data = tips,
          x_estimator = np.mean)
pass
```

```python
## With median values: 
sns.lmplot(x = 'size', y = 'tip', data = tips,
          x_estimator = np.median)
pass
```

# Fitting different kinds of models

* We consider `Anscombe's quartet` dataset shows a few examples where simple linear regression provides an identical estimate of a relationship where simple visual inspection shows differences. 

```python
anscombe = sns.load_dataset("anscombe")
anscombe = pd.DataFrame(data = anscombe)
anscombe.head()
```

```python
anscombe.columns
```

```python
sns.lmplot(x = 'x', y = 'y', 
           data = anscombe.query("dataset == 'I'"), 
          ci = None, scatter_kws = {"s": 80})
pass
```

```python
sns.lmplot(x = 'x', y = 'y', 
           data = anscombe.query("dataset == 'II'"), 
          ci = None, scatter_kws = {"s": 80})
pass
```

* It seems that we can fit a higher-order polynomial relations to explore simple kinds of nonlinear trends in the dataset. 

```python
sns.lmplot(x = 'x', y = 'y', 
           data = anscombe.query("dataset == 'II'"),
           order = 2,
          ci = None, scatter_kws = {"s": 80})
pass
```

```python
sns.lmplot(x = 'x', y = 'y', 
           data = anscombe.query("dataset == 'III'"), 
          ci = None, scatter_kws = {"s": 80})
pass
```

* There is one outlier observation in the dataset that deviates for some reason other than the main relationship under study. We can fit a robust regression which uses a different loss function to downweight relatively large residuals. 

```python
sns.lmplot(x = 'x', y = 'y', 
           data = anscombe.query("dataset == 'III'"),
           robust = True,
          ci = None, scatter_kws = {"s": 80})
pass
```

# Conditioning on other variables (categorical variable)

*  This is where the difference between regplot() and lmplot() appears. While regplot() always shows a single relationship, lmplot() combines regplot() with FacetGrid to provide an easy interface to show a linear regression on “faceted” plots that allow you to explore interactions with up to three additional categorical variables.

* The best way to separate out a relationship is to plot both levels on the same axes and to use color to distinguish them:

```python
sns.lmplot(x = 'total_bill', y = 'tip',
          hue = 'smoker', data = tips,
          ci = None)
pass
```

* In addition to color, it’s possible to use different scatterplot markers to make plots the reproduce to black and white better. You also have full control over the colors used:

```python
sns.lmplot(x = 'total_bill', y = 'tip',
          hue = 'smoker', data = tips,
          ci = None, 
          markers = ['o','x'], 
          palette = 'Set2')
pass
```

* To add another variable, you can draw multiple “facets” which each level of the variable appearing in the rows or columns of the grid:

```python
sns.lmplot(x = 'total_bill', y = 'tip',
          hue = 'smoker', col = 'time',data = tips,
          ci = None, 
          markers = ['o','x'], 
          palette = 'Set2')
pass
```

```python
sns.lmplot(x = 'total_bill', y = 'tip',
          hue = 'smoker', col = 'time',row = 'sex',
           data = tips,
          ci = None, 
          markers = ['o','x'], 
          palette = 'Set2')
pass
```
